## Dealing with Categorical Variables: Label Encoding and Hot Encoding

#### We used ordinal encoding for those variables that have a ranking for example: poor(1), fair(2) good(3), excelent(4), assigning a higher weight to those variables that have higher ranking.

#### We used one hot encoding for those variables that have a binary category such as 'yes or no', or 'wood or metal'

#### Finally we used purely categorical enconding with the function get_dummies

## Model 1: Linear Regression and Residuals

### Model 1 uses one-hot ecoded variables, ordinal encoded variables but excludes purely categorical variables

```python
from sklearn import datasets, linear_model, metrics
  
# defining feature matrix(X) and response vector(y)
X = data2.loc[:, data2.columns != 'SalePrice']
y = data2['SalePrice']
  
# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=1)
  
# create linear regression object
reg = linear_model.LinearRegression()
  
# train the model using the training sets
reg.fit(X_train, y_train)
  
# regression coefficients
#print('Coefficients: ', reg.coef_)
  
from sklearn.metrics import r2_score
r2 = r2_score(y, reg.predict(X))
print('R2 score: {}'.format(r2))


# plot for residual error
  
## setting plot style
plt.style.use('fivethirtyeight')
  
## plotting residual errors in training data
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train,
            color = "green", s = 10, label = 'Train data')
  
## plotting residual errors in test data
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test,
            color = "blue", s = 10, label = 'Test data')
  
## plotting line for zero residual error
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
  
## plotting legend
plt.legend(loc = 'upper right')
  
## plot title
plt.title("Residual Errors Model 1")
  
## method call for showing the plot
plt.show()
```

## Model 2: Model 1 +  purely categorical variables + data standarization

### Model 2 uses one-hot ecoded variables, ordinal encoded variables and purely categorical variables and data standarization

```python

from sklearn import datasets, linear_model, metrics
from sklearn.preprocessing import StandardScaler


#Data standarization
dummies1 = dummies.copy()
scaler = StandardScaler()


  
# defining feature matrix(X) and response vector(y)
X_1 = dummies1.loc[:, dummies.columns != 'SalePrice']
scaler.fit(X_1)
X_scaled = scaler.transform(X_1)

y_1 = dummies1['SalePrice']
  
# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.3,
                                                    random_state=1)
  
# create linear regression object
reg_1 = linear_model.LinearRegression()
  
# train the model using the training sets, taking log of response variable
reg_1.fit(X_train_1, np.log(y_train_1))

# regression coefficients
#print('Coefficients: ', reg_1.coef_)
  
# variance score: 1 means perfect prediction
from sklearn.metrics import r2_score
r2_1 = r2_score(y_1, np.exp(reg_1.predict(X_1)))
print('R2 score: {}'.format(r2_1))


# plot for residual error
  
## setting plot style
plt.style.use('fivethirtyeight')
  
## plotting residual errors in training data
plt.scatter(np.exp(reg_1.predict(X_train_1)), np.exp(reg_1.predict(X_train_1)) - y_train_1,
            color = "green", s = 10, label = 'Train data')
  
## plotting residual errors in test data
plt.scatter(np.exp(reg_1.predict(X_test_1)), np.exp(reg_1.predict(X_test_1)) - y_test_1,
            color = "blue", s = 10, label = 'Test data')
  
## plotting line for zero residual error
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
  
## plotting legend
plt.legend(loc = 'upper right')
  
## plot title
plt.title("Residual errors Model 2")
  
## method call for showing the plot
plt.show()
```
